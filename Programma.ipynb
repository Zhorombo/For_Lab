```python
from functools import lru_cache
import itertools
import collections

def win (a:list,k:int):                   #проверяем терминальное ли состояние
    p = 0
    for i in range (len(a)):
        p+= min(1,a[i])
    if p < k:
        return True
    return False

def stop(a:list):                   #есть ли следующий ход?
    for i in a:
        if i < 0:
            return False
    return True

def step(a:tuple,b:tuple):#переход к следующей комбинации a - текущая ситуация, b - ход
    d = list(a)
    for i in b:
        d[i]-=1
    return tuple(d)

def per (n,k):  #перебираем все возможные уникальные ходы
    return list(itertools.combinations(range(n),k))

@lru_cache
def mizer(a:tuple,k:int):
    res =list()
    n = len(a)
    if win(a,k):
        return 1
    else:
        for i in per(n,k):
            res.append(mizer(step(a,i),k))
        d = sorted(res,key = lambda x: (x%2,x))
        if (d[0]%2)==0:
            return d[0]+1
        else:
            return d[-1]+1
        
def cond(a:tuple):
    for i in range(len(a)-1):
        if a[i]>a[i+1]:
            return False
    return True

def condition(a:list):
    b = tuple(a)
    c = a.copy()
    d = sorted(c,key = lambda x: (x%2,x))
    if d[0]%2 == 1:
        return b.index(d[-1])
    return b.index(d[0]) 
```
